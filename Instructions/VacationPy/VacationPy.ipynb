### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Use google API key
gmaps.configure(api_key=g_key)

location = city_df[["Lat","Lng"]]
humidity = city_df["Humidity"]
```

```python
#Heatmap layer
figure_layout = {
    'width': '800px',
    'height': '500px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': 'auto'
}
heatmap = gmaps.figure(center = [0,0] ,zoom_level = 2., layout=figure_layout)

heat_layer = gmaps.heatmap_layer(location, weights=humidity,
                               dissipating=False, max_intensity=max(humidity),
                               point_radius = 2.8,)

heatmap.add_layer(heat_layer)
heatmap
```
